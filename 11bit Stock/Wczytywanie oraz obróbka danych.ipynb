## 1.Open
### Open oznacza cenę, po której rozpoczęto handel akcjami.
## 2.Close
### Close oznacza cenę, z którą kończą się akcje.
## 3.High
### High to najwyższa cena, po której handlowano akcjami w danym okresie.
## 4.Low
### Low to najniższa cena w danym okresie.
## 5.Volume
### Volume to całkowita liczba akcji będących przedmiotem obrotu w danym okresie. 
## 6.Adjusted close
### Adjusted close  to cena zamknięcia po korektach uwzględniających wszystkie stosowne podziały i wypłaty dywidendy.
